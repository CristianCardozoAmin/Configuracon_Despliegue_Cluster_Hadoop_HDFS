# Pruebas de funcionamiento e integración

Esta primera celda es un pequeño script en el cual podemos ver como funcionan y si estan funcinando los servicios de SPARK y HADOOP de foma integrada en este entorno de cluster.

```python
from pyspark.sql import SparkSession

# 1. Iniciar la sesión de Spark
print("Iniciando sesión de Spark...")
spark = SparkSession.builder \
    .appName("Test_Integracion_Spark_Hadoop") \
    .getOrCreate()

# Confirmar versiones
java_version = spark.sparkContext._jvm.java.lang.System.getProperty("java.version")
print(f"Versión de Spark: {spark.version}")
print(f"Versión de Java en uso: {java_version}")

# 2. Generar datos de prueba en la memoria de Spark
print("\n--- 1. Creando DataFrame en memoria (Prueba Spark) ---")
datos = [
    ("Hadoop", "Almacenamiento Distribuido HDFS", "OK"),
    ("Spark", "Procesamiento en Memoria", "OK"),
    ("Jupyter", "Interfaz Interactiva", "OK")
]
df = spark.createDataFrame(datos, ["Tecnologia", "Funcion", "Estado"])
df.show()

# 3. Escribir en Hadoop (HDFS)
print("\n--- 2. Escribiendo datos en el clúster (Prueba Hadoop) ---")
# HDFS tomará automáticamente la IP de tu NameNode gracias a la variable HADOOP_CONF_DIR
ruta_hdfs = "hdfs:///user/estudiante/test_integracion"

try:
    # Usamos overwrite para que puedas correr esta celda varias veces sin error
    df.write.mode("overwrite").parquet(ruta_hdfs)
    print(f"¡Éxito! Datos guardados físicamente en HDFS en: {ruta_hdfs}")
except Exception as e:
    print(f"Error al escribir en HDFS: {e}")

# 4. Leer desde Hadoop (HDFS)
print("\n--- 3. Leyendo datos desde el clúster ---")
try:
    df_leido = spark.read.parquet(ruta_hdfs)
    print("Datos recuperados correctamente de Hadoop:")
    df_leido.show()
    print("\n¡INTEGRACIÓN COMPLETADA CON ÉXITO! 🚀")
except Exception as e:
    print(f"Error al leer de HDFS: {e}")
```

### Comando 1: Creación de directorios
Genera un directorio de primer nivel para almacenar los datos.

`hdfs dfs -mkdir /folderName`

De igual manera, podemos utilizar la flag de `-p` para hacer recursivo el comando de creación de direcotrios, y asi crear en cadena varias carpetas

```python
!hdfs dfs -mkdir -p /folderA/experimentos/logs
!hdfs dfs -ls -R /folderA/
```

### Comando 5: Extraccion de un proceso
Permite extraer la salida de un proceso hacia el entorno local para su análisis final.

`hdfs dfs -get /sourceFile /outputPath`

```python
#Comando 5: hdfs dfs -get
!hdfs dfs -get /folderA/Prueba.txt /home/estudiante/Archivos-ingreso-hdfs/
```

(Descarga forzada sobreescribiendo): Sobreescribe cualquier archivo local que tenga el mismo nombre, para esto se usa la flag de `-f`

`hdfs dfs -get -f /sourceFile /outputPath`

```python
!hdfs dfs -get -f /folderA/Prueba1.txt /home/estudiante/Archivos-ingreso-hdfs
```

### Comando 6: Copiar datos
Mueve utilidades o datos de un espacio de trabajo a otro sin requerir descargarlos y volverlos a subir

`!hdfs dfs -cp /sourceFile /targetPath`

```python
#Comando 6: hdfs dfs -cp
!hdfs dfs -cp /folderA/Prueba.txt /folderB/
```

(Crear un respaldo interno de un directorio): Permite clonar un directorio completo dentro de HDFS garantizando un punto de restauración.

`!hdfs dfs -cp /folderA/file /folderB/backup`

```python
!hdfs dfs -cp /folderA/Prueba1.txt /folderB/backup
```

### Comando 7: limpieza de datos o folders
Elimina los datos o elementos de un directorio

`-r` (Recursivo): Por defecto, el comando rm en HDFS solo puede eliminar archivos individuales o carpetas que estén completamente vacías.

`-f` (Force / Forzar): Esta bandera es vital para la automatización en notebooks porque omitirá cualquier mensaje de advertencia que te pida confirmar si realmente deseas borrar los archivos e ignora los errores si el archivo o carpeta no existe.

`!hdfs dfs -rm /folder`

```python
# Comando para limpieza de todos los ejemplos ejecutados
!hdfs dfs -rm -r -f /folderA /folderB
```
